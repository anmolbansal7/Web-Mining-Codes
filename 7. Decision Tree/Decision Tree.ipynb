```python
#Anmol Bansal 19BCE0630
#importing the dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
```

```python
#importing the raw dataset

df = pd.read_csv("Iris.csv")
```

```python
#viewing the dataset

df
```

```python
#dropping useless columns and shuffling the dataset

df.drop("Id", axis = 1, inplace = True)
df = shuffle(df)
```

```python
#label encoding the output classes

cat = list(df['Species'].unique())
cat_label = dict(zip(cat, range(0, len(cat))))
df['Species'] = df['Species'].map(cat_label)
```

```python
#Viewing the preprocessed data

df
```

```python
#Seperating the dataset into training and testing set

X, y = df.drop("Species", axis = 1), df['Species']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, train_size = 0.66)
```

```python
#viewing training dataset

X_train
```

```python
y_train
```

```python
#viewing the test data

X_test
```

```python
y_test
```

```python
#Instantiating Decision Tree Object

dt = DecisionTreeClassifier()
```

```python
#fitting the model on training data

dt.fit(X_train, y_train)
```

```python
#predicting the model on test data

y_pred = dt.predict(X_test)
```

```python
#predicted values

y_pred
```

```python
#printing the classification report

print(classification_report(y_test, y_pred))
```

```python
#printing the confusion matrix

print(confusion_matrix(y_test, y_pred))
```

```python
#printing the accuracy of the dataset

print(dt.score(X_test, y_test))
```
